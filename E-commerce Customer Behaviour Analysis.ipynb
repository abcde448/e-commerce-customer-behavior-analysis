Q2: Find variance, standard deviation, and z-score (Purchase Amount)

```python
df.columns
```

```python
purchase_var = df["Purchase Amount ($)"].var()
purchase_std = df["Purchase Amount ($)"].std()
purchase_mean = df["Purchase Amount ($)"].mean()

z_scores = (df["Purchase Amount ($)"] - purchase_mean) / purchase_std

print("Variance:", purchase_var)
print("Standard Deviation:", purchase_std)
print("First 5 Z-scores:\n", z_scores.head())

```

The standard deviation in customer purchase amount is about 286.24 units, indicating how spread out spending is.The variance is 81932.240, confirming that spending habits vary significantly across users. Z-score helps identify which customers spend significantly more or less than the average

Q2: How do payment methods influence customer satisfaction and return rates?

```python
df_copy = df.copy()
satisfaction_map = {'Low': 1, 'Medium': 2, 'High': 3}
df_copy['Satisfaction_Score'] = df_copy['Customer Satisfaction'].map(satisfaction_map)

payment_group = df_copy.groupby('Payment Method').agg({
    'Satisfaction_Score': 'mean',
    'Return_Customer_Flag': 'mean',
    'Customer ID': 'count'
}).rename(columns={
    'Satisfaction_Score': 'Avg Satisfaction',
    'Return_Customer_Flag': 'Return Rate',
    'Customer ID': 'Total Customers'
}).sort_values(by='Avg Satisfaction', ascending=False)

print(payment_group)

```

```python
labels = payment_group.index.tolist()
satisfaction = payment_group['Avg Satisfaction'].values
return_rate = payment_group['Return Rate'].values

x = np.arange(len(labels))
width = 0.35
fig, ax = plt.subplots(figsize=(10, 6), facecolor='#121212')
ax.set_facecolor('#1e1e2f')
bars1 = ax.bar(x - width/2, satisfaction, width, label='Avg Satisfaction', color='#3498db', edgecolor='white')
bars2 = ax.bar(x + width/2, return_rate, width, label='Return Rate (%)', color='#2ecc71', edgecolor='white')
ax.set_title('Satisfaction & Return Rate by Payment Method', fontsize=15, fontweight='bold', color='white')
ax.set_xlabel('Payment Method', color='white')
ax.set_ylabel('Score / Rate', color='white')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=20, color='white')
ax.tick_params(axis='y', colors='white')
ax.legend(facecolor='#1e1e2f', edgecolor='white', labelcolor='white')
for bar1, bar2 in zip(bars1, bars2):
    height1 = bar1.get_height()
    ax.text(
        bar1.get_x() + bar1.get_width() / 2,
        height1 + 0.01,
        f"{height1:.2f}",
        ha='center',
        va='bottom',
        color='white',
        fontsize=10,
        fontweight='bold'
    )
    height2 = bar2.get_height()
    ax.text(
        bar2.get_x() + bar2.get_width() / 2,
        height2 + 0.01,
        f"{height2 * 100:.2f}%",
        ha='center',
        va='bottom',
        color='white',
        fontsize=10,
        fontweight='bold'
    )
ax.grid(color='gray', linestyle='--', linewidth=0.5, axis='y')

plt.tight_layout()
plt.savefig('level3_q2.png', dpi=300, bbox_inches='tight')
plt.show()

```

Among all payment methods, Bank Transfer and Credit Card users report the highest average satisfaction scores (≈2.01), with slightly above-average return rates. In contrast, Cash on Delivery users show the lowest satisfaction (1.96) and the lowest return likelihood (48.9%), indicating a possible friction in traditional payment experiences.
